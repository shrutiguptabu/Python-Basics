## Labeling the plots

```python
import numpy as np

x=np.linspace(0,10,30)
y=plt.plot(x, np.sin(x),'o', label='sin') # Label is the name of the legend


plt.title('beginners graph')
plt.xlabel('x axis') #to name the x axis
plt.ylabel('y axis') # to name the y axis
plt.axis([0,100,0,100]) 
plt.legend() # to plot the lagent
```

### Scatter plots

```python
%matplotlib inline   #  use matplotlib inline because output of plotting commands
#is displayed inline within frontends like the Jupyter notebook, directly below the code cell that produced it.
import matplotlib.pyplot as plt
# plt.style.use('seaborn-whitegrid')
import numpy as np
x = np.linspace(0, 100, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y,yerr=dy, fmt='or');
```

### Density and contour plots

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x=np.linspace(0,10,50)
# y=np.linspace(0,10,50)
# z=np.linspace(0,10,50)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
```

```python
plt.contour(X,Y,Z)
plt.colorbar()
```

### Customizing colorbars

```python
x = np.linspace(0, 10, 1000)
I = np.sin(x) * np.cos(x[:, np.newaxis])

plt.imshow(I)
plt.colorbar();
```
